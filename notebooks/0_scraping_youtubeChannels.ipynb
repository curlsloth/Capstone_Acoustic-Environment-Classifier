```python
def getChanVideoList(channel_id):
    playlist = Playlist(playlist_from_channel_id(channel_id))

    print(f'Videos Retrieved: {len(playlist.videos)}')

    while playlist.hasMoreVideos:
        print('Getting more videos...')
        playlist.getNextVideos()
        print(f'Videos Retrieved: {len(playlist.videos)}')

    print('Found all the videos.')
    return pd.DataFrame(playlist.videos)
```

```python
playlist = Playlist('https://www.youtube.com/watch?v=ECSIkIC6RZw&list=PLkmb1BBi8y93qIwNcnuo3e9oyLxc8PVGB')

print(f'Videos Retrieved: {len(playlist.videos)}')

while playlist.hasMoreVideos:
    print('Getting more videos...')
    playlist.getNextVideos()
    print(f'Videos Retrieved: {len(playlist.videos)}')

print('Found all the videos.')
NomadicAmbience_nature_df = pd.DataFrame(playlist.videos)
```

```python
from datetime import datetime
DATE_FORMATS = ['%H:%M:%S', '%M:%S']
issues = set()
def parse(x_):
    for fmt in DATE_FORMATS:
        try:
            return datetime.strptime(x_, fmt).strftime('%H:%M:%S')
        except ValueError:
            pass
    issues.add(x_)


nature_df['duration_sec'] = pd.to_timedelta(nature_df['duration'].apply(lambda x: parse(x))).apply(lambda x: x.total_seconds())
city_df['duration_sec'] = pd.to_timedelta(city_df['duration'].apply(lambda x: parse(x))).apply(lambda x: x.total_seconds())
NomadicAmbience_nature_df['duration_sec'] = pd.to_timedelta(NomadicAmbience_nature_df['duration'].apply(lambda x: parse(x))).apply(lambda x: x.total_seconds())

# drop the videos < 120 seconds
nature_df = nature_df[nature_df['duration_sec']>120]
city_df = city_df[city_df['duration_sec']>120]
NomadicAmbience_nature_df = NomadicAmbience_nature_df[NomadicAmbience_nature_df['duration_sec']>120]
```

# Download youtube as audio files

```python
# https://github.com/ytdl-org/youtube-dl
# https://github.com/yt-dlp/yt-dlp

# from __future__ import unicode_literals
# import youtube_dl

from yt_dlp import YoutubeDL
from yt_dlp.utils import download_range_func

class MyLogger(object):
    def debug(self, msg):
        pass

    def warning(self, msg):
        pass

    def error(self, msg):
        print(msg)


def my_hook(d):
    if d['status'] == 'finished':
        print('Done downloading, now converting ...')


ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'm4a',
        'preferredquality': '192',
    }],
    'ffmpeg_location': '/Users/andrewchang/opt/anaconda3/envs/Springboard_GoogleAudio/bin/ffmpeg',
    'logger': MyLogger(),
    'progress_hooks': [my_hook],
}
# with youtube_dl.YoutubeDL(ydl_opts) as ydl:
#     ydl.download(['https://www.youtube.com/watch?v=BaW_jenozKc'])
```

```python
import warnings
warnings.filterwarnings("ignore")

ydl_opts['outtmpl'] = '/Users/andrewchang/Springboard_GitHub/Capstone_AcousticEnv/data/raw/youtube/NomadicAmbience_nature/%(id)s_%(start_time)i.m4a'

import random
random.seed(23)

with YoutubeDL(ydl_opts) as ydl:
    for index, row in NomadicAmbience_nature_df.iterrows():
        nRep = 0
        while nRep < 1: # number of segments within the same file
            dl_start_sec = random.randint(60,row['duration_sec']-15)
            ydl_opts['download_ranges'] = download_range_func(None, [(dl_start_sec,dl_start_sec+10)])
            ydl.download('https://www.youtube.com/embed/'+row['id']+'?start='+str(dl_start_sec))
            nRep += 1
```
