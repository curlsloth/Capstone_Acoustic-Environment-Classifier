```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
meta_df = pd.read_csv('../data/interim/train_val_test_split_Feb21.csv', index_col=0)
meta_df
```

```python
def load_df_pkl(filename):
    import pickle
    # open a file, where you stored the pickled data
    file = open(filename, 'rb')

    # dump information to that file
    output = pickle.load(file)
    df_indices = output['df_indices']
#     df_indices_fg = output['df_indices_fg']
#     df_indices_bg = output['df_indices_bg']
#     df = pd.concat([df_indices.add_suffix('_raw'), df_indices_fg.add_suffix('_fg'), df_indices_bg.add_suffix('_bg')], axis=1)
    df = df_indices

    # close the file
    file.close()
    return df

df_list = []
for index, row in meta_df.iterrows():
    df_list.append(load_df_pkl(row['file']))

df = pd.concat(df_list, ignore_index=True)
df = pd.concat([df, meta_df[['category','fold']]], axis=1)
df
```

```python
# replacing the inf values by NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
train_df = df[df['fold']<8].drop(columns='fold')
val_df = df[df['fold']==8].drop(columns='fold')
test_df = df[df['fold']==9].drop(columns='fold')
```

```python
# replace the nan by median
from sklearn.impute import SimpleImputer
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
imp_median.fit(train_df)
train_df = pd.DataFrame(imp_median.transform(train_df), columns=train_df.columns)
val_df = pd.DataFrame(imp_median.transform(val_df), columns=val_df.columns)
test_df = pd.DataFrame(imp_median.transform(test_df), columns=test_df.columns)
```

```python
# import the necessary packages
import tensorflow as tf
tf.random.set_seed(23)

tf.keras.backend.clear_session()

# define our MLP network
# model = tf.keras.Sequential([
#     tf.keras.layers.Input(shape=(180), dtype=tf.float32),
#     tf.keras.layers.Dense(180, activation='relu'),
#     tf.keras.layers.Dense(180, activation='relu'),
#     tf.keras.layers.Dense(90, activation='relu'),
#     tf.keras.layers.Dense(90, activation='relu'),
#     tf.keras.layers.Dense(90, activation='relu'),
#     tf.keras.layers.Dense(2)
# ], name='model')

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(60), dtype=tf.float32),
#     tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(2)
], name='model')

model.summary()
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                    optimizer="adam",
                    metrics=['accuracy'])

callback = tf.keras.callbacks.EarlyStopping(monitor='loss',
                                            patience=3,
                                            restore_best_weights=True)


history = model.fit(x=train_df.drop(columns='category'),
                    y=train_df['category'],
                    epochs=20,
                    validation_data=(val_df.drop(columns='category'), val_df['category']),
                    callbacks=callback)
```

```python
loss, accuracy = model.evaluate(x=test_df.drop(columns='category'), y=test_df['category'])

print("Loss: ", loss)
print("Accuracy: ", accuracy)
```
